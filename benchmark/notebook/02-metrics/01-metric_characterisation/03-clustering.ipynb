```python
import cfe
import matplotlib.pyplot as plt
import scanpy as sc
```

绘制聚类结果

```python
fadata = cfe.data.FateAnnData.read_dynverse_simulation_data()

fig, ax = plt.subplots(1, 3, figsize=(3 * 5, 5), dpi=120)
cfe.plot.plot_graph(fadata, ax=ax[0])

cluster_key = "milestone_color"
fadata.group_onto_nearest_milestones(cluster_key=cluster_key) # new cluster color
cfe.plot.plot_graph(fadata, color=cluster_key, ax=ax[1])

cluster_key = "edge_color"
fadata.group_onto_trajectory_edges(cluster_key=cluster_key) # new cluster color
cfe.plot.plot_graph(fadata, color=cluster_key, ax=ax[2])
```
